Dana jest prostokątna, [rzadka](https://en.wikipedia.org/wiki/Sparse_matrix) macierz $\mathbf{A}$ o wymiarach $m \times n$ oraz wektor kolumnowy $\mathbf{b}$ o wymiarach $m \times 1$. Należy rozwiązać układ równań postaci:
 
$$
\mathbf{Ax}=\mathbf{b}
$$
  
gdzie:
- $\mathbf{A}$ - macierz współczynników z lewej strony równania, 
- $\mathbf{x}$ - wektor kolumnowy zmiennych,
- $\mathbf{b}$ - wektor kolumnowy wyników prawej strony równania. 

Do oceny jakości rozwiązania wykorzystywać będziemy residuum (ang. *residual*):

$$
\mathbf{r}=\mathbf{b}-\mathbf{Ax}
$$

```python
# Fragment kodu wyznaczający ilość zużycia pamięci przez macierze (w bajtach).
# n: np.ndarray = 
# n_memory_size = n.nbytes

# s: sp.sparse.csc_array = ...
# s_memory_size = s.data.nbytes + s.indices.nbytes + s.indptr.nbytes
```

```python
# ===== Twoja implementacja tutaj =====
m = np.linspace(100,10000,10).astype(int)
matrix = []
matrix_np = []
memory_mat = []
memory_np = []
for idx,param in enumerate(m):
    matrix.append(sp.sparse.random(param, param, density=0.01, format="csc"))
    matrix_np.append(matrix[idx].toarray())
    size_s = matrix[idx].data.nbytes + matrix[idx].indices.nbytes + matrix[idx].indptr.nbytes
    memory_mat.append(size_s)
    size_d = matrix_np[idx].nbytes
    memory_np.append(size_d)
plt.scatter(m, memory_mat, label="macierz rzadka", color="red")
plt.scatter(m, memory_np, label="macierz gęsta", color="blue")
plt.yscale('log')
plt.legend()
plt.show()
```

Wykres obrazuje zależność zajętości pamięci operacyjnej w bajtach od rozmiaru macierzy m dla dwóch formatów zapisu: gęstego (numpy.ndarray) oraz rzadkiego (scipy.sparse.csc). Ze względu na różnice rzędów wielkości zastosowano skalę logarytmiczną na osi Y. Z analizy wykresu wynika, że macierz gęsta, która alokuje pamięć dla wszystkich elementów niezależnie od ich wartości, wykazuje zapotrzebowanie na zasoby rosnące proporcjonalnie do kwadratu wymiaru macierzy. Widać to po znacznie wyższych wartościach dla formatu gęstego (punkty górne). W przeciwieństwie do niej, macierz rzadka przechowuje w pamięci jedynie wartości niezerowe oraz ich indeksy, co przy ustalonej gęstości (density=0.01) skutkuje drastycznie mniejszym zużyciem pamięci RAM (punkty dolne). Potwierdza to konieczność stosowania formatów rzadkich w obliczeniach numerycznych operujących na dużych macierzach o niskim stopniu wypełnienia, aby uniknąć przepełnienia pamięci komputera.

**Punkt 3.**

Istnieje wiele metod iteracyjnego rozwiązywania równań, natomiast nie różnią się one znacząco od siebie w kwestii wywołania metody, dlatego w dalszej części zadania należy wybrać jedną z metod: [GMRES](https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.linalg.gmres.html) lub [Conjugate Gradient](https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.linalg.cg.html) i dla niej przeprowadzić dalszą analizę. 

1. Korzystając z funkcji *`residual_norm()`*, oraz macierzy zdefiniowanych w **Punkcie 2.**, zbadaj wpływ argumentów `rtol` i `maxiter` na działanie wybranej metody. Sporządź wykresy przedstawiające wartość normy residuum w zależności od zmiany podanych parametrów.

2. Opisz zaobserwowane zależności.

>**Wskazówka:** Metoda *Conjugate Gradient* wymaga aby macierz była dodatnio określona. Można to zapewnić poprzez podanie argumentu `is_symmetric=True` do funkcji `generate_matrix()`.


```python
# Implementacja funkcji generate_matrix().

from scipy.sparse import diags_array, csc_array


def generate_matrix(m: int = 1000, is_symmetric: bool = False) -> csc_array:
    """Funkcja generująca rzadką, diagonalnie zdominowaną macierz rozmiaru m x m.

    Parameters:
        m (int): Wymiar macierzy.
        is_symmetric (bool): Czy macierz ma być symetryczna.

    Returns:
        (scipy.sparse.csc_array): Macierz rzadka, diagonalnie zdominowana (m,m).

    Raises:
        TypeError: Jeśli 'm' nie jest liczbą całkowitą lub 'is_symmetric' nie 
            jest wartością logiczną.
        ValueError: Jeśli 'm' jest mniejsze niż 5.
    """
    if not isinstance(m, int):
        raise TypeError(
            f"Argument 'm' musi być liczbą całkowitą, otrzymano: {type(m).__name__}."
        )
    if not isinstance(is_symmetric, bool):
        raise TypeError(
            f"Argument 'is_symmetric' musi być wartością logiczną, otrzymano: {type(is_symmetric).__name__}."
        )
    if m < 5:
        raise ValueError("Argument 'm' musi być większy lub równy 5.")

    if is_symmetric:
        num_upper_diags = np.random.randint(2, m // 2)
        diagonal_values = np.random.rand(num_upper_diags)
        matrix = diags_array(
            diagonal_values, offsets=np.arange(1, num_upper_diags + 1), shape=(m, m)
        )
        matrix = matrix + matrix.T
    else:
        num_upper_diags = np.random.randint(2, m // 2)
        num_lower_diags = np.random.randint(2, m // 2)
        diagonal_values = np.random.rand(num_upper_diags + num_lower_diags)
        matrix = diags_array(
            diagonal_values,
            offsets=np.arange(-num_lower_diags, num_upper_diags),
            shape=(m, m),
        )

    row_sums = np.asarray(matrix.sum(axis=1)).flatten()
    diagonal_vals = row_sums - np.abs(matrix.diagonal()) + np.random.rand(m)
    matrix.setdiag(diagonal_vals)

    return matrix.tocsc()
```

```python
def is_diagonally_dominant(A: np.ndarray | sp.sparse.csc_array) -> bool | None:
    if not isinstance(A, np.ndarray | sp.sparse.csc_array):
        return None
    if sp.sparse.issparse(A):
        A = A.toarray()
    if A.ndim != 2:
        return None
    if A.shape[0] != A.shape[1]:
        return None
    abs_A = np.abs(A)
    diagonal = np.diag(abs_A)
    sum_rows = np.sum(abs_A, axis=1)
    sum_without_diag = sum_rows - diagonal
    
    return np.all(diagonal > sum_without_diag)

def residual_norm(A: np.ndarray, x: np.ndarray, b: np.ndarray) -> float | None:
    if not isinstance(A, np.ndarray) or not isinstance(x, np.ndarray) or not isinstance(b, np.ndarray):
        return None
    if A.ndim != 2:
        return None
    if A.shape[0] != b.shape[0]:
        return None
    if A.shape[1] != x.shape[0]:
        return None
    if b.ndim != 1 or x.ndim != 1:
        return None
    return float(np.linalg.norm(b - A @ x))

# ===== Twoja implementacja tutaj =====
A = generate_matrix(1000)
print(is_diagonally_dominant(A))
B = np.random.randint(0, 100, 1000)
A = A.toarray()

rtol_values = [1e-1, 1e-3, 1e-5, 1e-7, 1e-9, 1e-11, 1e-13, 1e-15]
fixed_maxiter = 3000 

gmres_rtol_norms = []

for tol in rtol_values:
    x_sol, info = sp.sparse.linalg.gmres(A, B, rtol=tol, maxiter=fixed_maxiter, atol=0)
    norm = residual_norm(A, x_sol, B)
    gmres_rtol_norms.append(norm)


maxiter_values = range(1, 50) 
fixed_rtol = 1e-16

gmres_iter_norms = []

for m_iter in maxiter_values:
    x_sol, info = sp.sparse.linalg.gmres(A, B, rtol=fixed_rtol, maxiter=m_iter, restart=1, atol=0)
    norm = residual_norm(A, x_sol, B)
    gmres_iter_norms.append(norm)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# WYKRES 1: Zależność od RTOL
ax1.plot(rtol_values, gmres_rtol_norms, marker='o', color='blue', label='GMRES')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_title(f'Wpływ parametru rtol\n(przy maxiter={fixed_maxiter})')
ax1.set_xlabel('Wartość rtol (oczekiwana dokładność)')
ax1.set_ylabel('Rzeczywista norma residuum')
ax1.invert_xaxis() 
ax1.grid(True, which="both", ls="--")
ax1.set_ylim(bottom=1e-17) 

# WYKRES 2: Zależność od MAXITER
ax2.plot(maxiter_values, gmres_iter_norms, marker='s', color='red', label='GMRES')
ax2.set_yscale('log')
ax2.set_title(f'Wpływ parametru maxiter\n(przy rtol={fixed_rtol})')
ax2.set_xlabel('Liczba iteracji (maxiter)')
ax2.set_ylabel('Norma residuum')
ax2.grid(True, which="both", ls="--")
ax2.set_ylim(bottom=1e-17)

plt.tight_layout()
plt.show()
```

Przedstawione wykresy obrazują analizę wpływu parametrów sterujących (rtol oraz maxiter) na dokładność rozwiązania układu równań metodą GMRES. Na lewym wykresie widoczna jest liniowa zależność w skali logarytmicznej między zadaną tolerancją a otrzymaną normą residuum, co potwierdza, że algorytm precyzyjnie realizuje zadany warunek stopu, kończąc obliczenia w momencie osiągnięcia wymaganej dokładności. Prawy wykres przedstawia charakterystykę zbieżności metody, gdzie błąd maleje wykładniczo z każdą kolejną iteracją (przy wymuszonym restarcie), nie wykazując przy tym stagnacji ani niestabilności. Tak modelowe zachowanie i wysoka efektywność metody wynikają bezpośrednio z własności badanej macierzy, która jest silnie diagonalnie zdominowana, a tym samym bardzo dobrze uwarunkowana numerycznie.
